```python
import pandas as pd
df=pd.read_csv("Hospital_patient_details.csv")
df.tail()
df.columns
df.info()
df.describe()
df.isnull().sum()
```

```python
import pandas as pd
def patient_category(days):
  if days >=8:
    return 'Crtical'
  elif days >=5:
      return 'Moderate'
  else:
    return 'Normal'
df['Patient_Category']=df['Length_of_Stay'].apply(patient_category)
print(df[['Patient_ID',"Disease",'Length_of_Stay','Patient_Category']].tail(7))
```

```python
import matplotlib.pyplot as plt

disease_count = df['Disease'].value_counts()

disease_count.plot(
    kind='bar',
    color=['green','blue','orange','yellow','red','purple','cyan']
)
plt.title('Patient Count by Disease')
plt.xlabel('Disease')
plt.ylabel('Number of Patients')
plt.show()
```

```python
age_group_count=df['Age_Group'].value_counts()
age_group_count.plot(
    kind='bar',
    color=['skyblue','orange','green','red']
)
plt.title('Patient Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Disease'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Disease Distribution')
plt.ylabel('')
plt.show()
```

```python
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])

df['Month'] = df['Admission_Date'].dt.month_name()

monthly_admissions = df['Month'].value_counts()

monthly_admissions.plot(
    kind='bar',
    color='teal'
)

plt.title('Monthly Patient Admissions')
plt.xlabel('Month')
plt.ylabel('Number of Admissions')
plt.show()
```

```python
# Total Patients
print("Total Patients:", len(df))

# Disease Wise Count
print(df['Disease'].value_counts())

# Age Group Wise Count
print(df['Age_Group'].value_counts())

# Average Length of Stay
print("Average Length of Stay:",
      df['Length_of_Stay'].mean())
```
